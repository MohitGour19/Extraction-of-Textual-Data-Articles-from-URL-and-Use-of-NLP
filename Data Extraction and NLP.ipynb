```python
import pyphen
dic = pyphen.Pyphen(lang='en')
def count_syllables(word):
    
    syllables = dic.inserted(word).count('-') + 1
    return syllables
```

```python
def complex_word_count(row):
    text=row.split()
    count=0
    for word in text:
        if count_syllables(word)>2:
            count+=1
    return count
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
def count_words_and_sentences(text):
    words = word_tokenize(text)
    sentences = sent_tokenize(text)
    word_count = len(words)
    sentence_count = len(sentences)
    return word_count/sentence_count
```

```python

def total_word_nltk(row):
    text = row['Text']

    words = word_tokenize(text)
    
    pattern = r'[^\w\s]'
    
    word_count = 0
    for word in words:
        cleaned_word = re.sub(pattern, '', word)
        if cleaned_word.lower() not in set(stopwords.words('english')):
            word_count += 1
    
    return word_count
```

```python
def Personal_Pronouns(row):
    text = row['Text']
    pattern = r'\b(?:I|we|my|ours|us)\b'
    matches = re.findall(pattern, text, flags=re.IGNORECASE)

     # Filter out matches that are the country name "US"
    personal_pronouns = [match for match in matches if match.lower() != 'us']
    
    personal_count = len(personal_pronouns)
    
    
    return personal_count
```

```python
Average Word Length is calculated by the formula:
Sum of the total number of characters in each word/Total number of words
```

```python
def average_word_length(row):
    text = row['Text']
    words = word_tokenize(text)
    count_word = 0
    count_char = 0;
    for word in words:
        count_word+=1
        for char in word:
            count_char+=1
            
    if(count_word==0):
        return 0
    else:
         return (count_char/count_word) 
```

```python
scraped_df['Average Word Length'] = scraped_df.apply(average_word_length,axis=1)
```

```python
Syllable Count Per Word
We count the number of Syllables in each word of the text by counting the vowels present in each word. We also handle some exceptions like words ending with "es","ed" by not counting them as a syllable.
```

```python
def syllable_count_per_word(row):
    text = row['Text']
    words = word_tokenize(text)
    count_syllable = 0
    for word in words:
        
        word = word.lower()
        vowels = re.findall(r'[aeiouy]+', word)
    
        if word.endswith(('es', 'ed')) and len(vowels) > 1:
             count_syllable+=len(vowels) - 1
        else:
            count_syllable+=len(vowels)
        
    count_word = 0
    for word in words:
        count_word+=1
    
    if count_word==0:
        return 0
    else:
        return count_syllable/count_word
```

```python
scraped_df['Syllable Count Per Word'] = scraped_df.apply(syllable_count_per_word,axis = 1)
```
